##**Objective:**

- WAP to evaluate the performance of implemented three-layer neural network with variations in activation functions, size of hidden layer, learning rate, batch size and number of epochs.

##**Description of the Model:**

This is a three-layer neural network implemented using TensorFlow (without Keras) for classifying handwritten digits from the MNIST dataset.The model consists of multiple layers and is trained using Mini-batch Gradient Descent with different batch sizes and epoch configurations.

The model consists of:

Input Layer (784 neurons): Accepts the 1D vector of 784 pixel values.

Hidden Layer 1 (128 neurons): Uses ReLU activation to learn non-linear features.

Hidden Layer 2 (64 neurons): Uses ReLU activation to extract deeper representations.

Output Layer (10 neurons): Uses softmax activation (integrated in the loss function) to classify the digit.

Loss Function: Softmax Cross-Entropy.

Optimizer: Adam Optimizer for efficient learning.

Training Method: Mini-batch Gradient Descent with different batch sizes.

- Epochs: Varies (100, 50, 10) to analyze model performance.
- Batch Sizes: 100, 10, and 1 (affects speed and accuracy).


##**Python Implementation**

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import time
import seaborn as sns
from sklearn.metrics import confusion_matrix
from tensorflow.python.framework import ops

# Disable eager execution
tf.compat.v1.disable_eager_execution()

# Load MNIST dataset
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Normalize and reshape input data
x_train, x_test = x_train / 255.0, x_test / 255.0
x_train = x_train.reshape(-1, 784)
x_test = x_test.reshape(-1, 784)

# One-hot encoding for labels
y_train_one_hot = np.eye(10)[y_train]
y_test_one_hot = np.eye(10)[y_test]

# Function to initialize weights
def init_weights(shape):
    return tf.Variable(tf.random.normal(shape, stddev=0.1))

# Define placeholders
X = tf.compat.v1.placeholder(tf.float32, [None, 784])
Y = tf.compat.v1.placeholder(tf.float32, [None, 10])

# Define model parameters
W1 = init_weights([784, 128])
b1 = tf.Variable(tf.zeros([128]))
W2 = init_weights([128, 64])
b2 = tf.Variable(tf.zeros([64]))
W3 = init_weights([64, 10])
b3 = tf.Variable(tf.zeros([10]))

# Forward propagation
def forward_propagation(X):
    z1 = tf.matmul(X, W1) + b1
    a1 = tf.nn.relu(z1)
    z2 = tf.matmul(a1, W2) + b2
    a2 = tf.nn.relu(z2)
    z3 = tf.matmul(a2, W3) + b3
    return z3  # No softmax since it's included in loss function

y_logits = forward_propagation(X)
# Loss function
#loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(logits=y_pred, labels=Y))
# Change from:
# loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(logits=y_pred, labels=Y))
# To:
loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=y_logits, labels=Y))
optimizer = tf.compat.v1.train.AdamOptimizer().minimize(loss)

# Accuracy calculation
correct_pred = tf.equal(tf.argmax(y_logits, 1), tf.argmax(Y, 1))
accuracy = tf.reduce_mean(tf.cast(correct_pred, tf.float32))

# Training configurations
batch_sizes = [100, 10, 1]
epochs_list = [100, 50, 10]

results = {}

# Run training for different configurations
for batch_size in batch_sizes:
    for epochs in epochs_list:
        print(f"\nTraining with Batch Size={batch_size}, Epochs={epochs}")
        train_losses = []
        train_accuracies = []
        start_time = time.time()

        with tf.compat.v1.Session() as sess:
            sess.run(tf.compat.v1.global_variables_initializer())
            for epoch in range(epochs):
                for i in range(0, x_train.shape[0], batch_size):
                    batch_x = x_train[i:i+batch_size]
                    batch_y = y_train_one_hot[i:i+batch_size]
                    sess.run(optimizer, feed_dict={X: batch_x, Y: batch_y})

                train_loss, train_acc = sess.run([loss, accuracy], feed_dict={X: x_train, Y: y_train_one_hot})
                train_losses.append(train_loss)
                train_accuracies.append(train_acc)
                print(f"Epoch {epoch+1}/{epochs}, Loss: {train_loss:.4f}, Accuracy: {train_acc:.4f}")

            test_acc, y_pred_logits = sess.run([accuracy, y_logits], feed_dict={X: x_test, Y: y_test_one_hot})
            execution_time = time.time() - start_time

            # Compute confusion matrix
            y_pred_classes = np.argmax(y_pred_logits, axis=1)
            conf_matrix = confusion_matrix(y_test, y_pred_classes)

            results[(batch_size, epochs)] = {
                "train_losses": train_losses,
                "train_accuracies": train_accuracies,
                "test_accuracy": test_acc,
                "execution_time": execution_time,
                "conf_matrix": conf_matrix
            }

            print(f"Test Accuracy: {test_acc:.4f}, Execution Time: {execution_time:.2f} sec")

        # Plot Loss and Accuracy curves
        plt.figure(figsize=(12, 5))

        plt.subplot(1, 2, 1)
        plt.plot(range(1, epochs + 1), train_losses, label=f"Batch={batch_size}, Epochs={epochs}")
        plt.xlabel("Epochs")
        plt.ylabel("Loss")
        plt.title("Loss Curve")
        plt.legend()

        plt.subplot(1, 2, 2)
        plt.plot(range(1, epochs + 1), train_accuracies, label=f"Batch={batch_size}, Epochs={epochs}")
        plt.xlabel("Epochs")
        plt.ylabel("Accuracy")
        plt.title("Accuracy Curve")
        plt.legend()

        plt.show()

        # Plot Confusion Matrix
        plt.figure(figsize=(8, 6))
        sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
        plt.xlabel("Predicted Label")
        plt.ylabel("True Label")
        plt.title(f"Confusion Matrix (Batch={batch_size}, Epochs={epochs})")
        plt.show()

# Print Execution Time Comparison
print("\nExecution Time Comparison:")
for key, val in results.items():
    print(f"Batch Size={key[0]}, Epochs={key[1]} -> Execution Time: {val['execution_time']:.2f} sec")
```

##**Code Explanation:**

1. Load & Preprocess Data
- The MNIST dataset is loaded using tf.keras.datasets.mnist.load_data(), which provides 60,000 training and 10,000 testing grayscale images of handwritten digits (28×28 pixels).

- Images are normalized by dividing by 255.0 to scale pixel values into the range [0,1].
- Images are flattened from 28×28 matrices to 1D vectors of 784 elements (x_train.reshape(-1, 784)).
- Labels are converted to one-hot encoding (np.eye(10)[y_train]), converting class labels (0-9) into a 10-element binary vector.

2. Initialize Model Parameters
- Weights (W1, W2, W3) initialized with small random values.
- Biases (b1, b2, b3) initialized as zeros.

3. Feed-Forward Propagation
- Layer 1: a1 = ReLU(X * W1 + b1).
- Layer 2: a2 = ReLU(a1 * W2 + b2).
- Output Layer: softmax(a2 * W3 + b3).

4. Loss Calculation
- Uses softmax_cross_entropy_with_logits_v2.

5. Backpropagation & Optimization
- Uses AdamOptimizer().
- Updates weights and biases.
6. Training (Mini-Batch Gradient Descent)

- Three different batch sizes (100, 10, 1) and epochs (100, 50, 10) are tested.

- Training loop:
   - For each epoch, data is split into mini-batches.
   - Training loss and accuracy are computed at the end of each epoch.

7. Testing & Evaluation
- After training, model performance is evaluated on the test dataset.
- Confusion Matrix:
   - The confusion matrix is computed using confusion_matrix(y_test, y_pred_classes).
   - A heatmap visualization is generated with seaborn.heatmap()


8. Performance Visualization
- Loss & Accuracy Curves:
   - Plots training loss & accuracy over epochs.
   - plt.plot(range(1, epochs + 1), train_losses, label="...")
- Confusion Matrix Heatmap:
  - Visualizes model predictions.
